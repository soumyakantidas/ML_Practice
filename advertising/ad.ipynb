# Advertising Data

** Python imports **

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ad = pd.read_csv('Advertising.csv')
```

Renamed unnamed column as *Market* and set it as index

```python
ad.rename(columns= {'Unnamed: 0': 'Market'}, inplace=True)
ad.set_index('Market', inplace=True)
ad.head()
```

```python
ad.describe()
```

```python
sns.pairplot(ad)
```

```python
sns.heatmap(ad.corr(), annot=True, cmap='Greens')
```

** From the pairplot and heatmap it is clear that TV ad budget is highly correlated with Sales **

```python
sns.set_style('whitegrid')
sns.lmplot(x='TV', y='Sales', data=ad, size=6, aspect= 1.5)
```

```python
sns.lmplot(x='Radio', y='Sales', data=ad, size=6, aspect= 1.5)
```

```python
sns.lmplot(x='Newspaper', y='Sales', data=ad, size=6, aspect= 1.5)
```
